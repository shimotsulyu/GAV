```python
#Calculate the r square
def rsquare(x,z,degree=1):
    coeffs = np.polyfit(x,z,degree)
    p = np.poly1d(coeffs)
    zhat = p(x)
    zbar = np.sum(z)/len(z)
    ssreg = np.sum((zhat-zbar)**2)
    sstot = np.sum((z-zbar)**2)
    r2 = ssreg/sstot
    #r2 pontos na reta
    #coeffs[0] = a iclination
    #coeffs[1] = b
    if degree==1:
        return r2, coeffs[0]
    elif degree==2:
        return r2, coeffs[1]
    else:
        return r2, coeffs
    
```

```python
#coleta os dados na direção de inclinação theta
import numpy as np
TAM, tam= 100, 5 #tamanho da matriz
data = np.random.randint(100, size=(TAM,TAM)) #cria matriz aleatoria
print('\ndata:\n',data)
```

```python
#equação da reta y = tan(thetta)*x+b
erro, point, theta = 1, [0,0], 45 #ponto de referencia e angulo de inclinação
b = point[1]-round(np.tan(np.radians(theta)),2)*point[0]
z = np.zeros(tam*3)
x = np.zeros(tam*3)
count = 0
for i in range(0,tam):
     for j in range(0,tam):
            if b-erro<=j-round(np.tan(np.radians(theta)),2)*i and j-round(np.tan(np.radians(theta)),2)*i<=b+erro:
                x[count] = round(np.cos(np.radians(theta))*i + np.sin(np.radians(theta))*j,2)
                z[count] = data[i,j]
                count+=1

z = z[0:np.count_nonzero(z)]
x = x[0:len(z)]
print('z:\n',z)
print('x:\n',x)
print('len',len(z))
r2, coeffs = rsquare(x,z)
print('r2',r2)
print('coeffs',coeffs)
print('----------------------------------------------------------------------------------')
```
